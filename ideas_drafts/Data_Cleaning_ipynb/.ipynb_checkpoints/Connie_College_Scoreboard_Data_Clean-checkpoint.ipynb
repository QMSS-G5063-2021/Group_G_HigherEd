## Cleaning Data - Removal of For-Profit and Community Colleges

We can start by removing for-profit universities. 

```python
#removal of for profit universities 
df = df[df['CONTROL']!=3]
```

```python
import numpy as np 

plt.hist(df.SAT_AVG, color ='#91B9D6')
plt.title("SAT Distribution")
plt.show()  
```

```python
plt.hist(df.ACTCMMID, color ='#897C90')
plt.title('ACT Distribution (Valid Records Only)')
```

```python
print('Note: Records without the field populated are denoted as 0. Records with the field populated are denoted as 1.')
```

```python
df['SAT_AVG_1'] = df['SAT_AVG'].fillna(0)
df['SAT_AVG_1'] = df['SAT_AVG_1'].apply(lambda x:1 if x!=0 else 0)

top_Instituition = (df['SAT_AVG_1'].value_counts()*100/4108)
ax = top_Instituition[0:2].plot(kind='barh', color ='#91B9D6')
plt.title('% Matched Records - SAT')
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
```

```python
df['ACTCMMID'] = df['ACTCMMID'].fillna(0)
df['ACT_1'] = df['ACTCMMID'].apply(lambda x:1 if x!=0 else 0)

top_Instituition = (df['ACT_1'].value_counts()*100/4108)
ax = top_Instituition[0:2].plot(kind='barh', color ='#897C90')
plt.xlabel('No. of Individuals')
plt.title('Top Unmatched Records - ACT')
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
```

Lets can see if we can improve the number of n/a records. 

```python
#now we remove the community colleges. 
#df2 is our listing of community colleges. 
df2 = df2.rename(columns={'UNIT_ID ':'UNITID'})
df['UNITID'] = df['UNITID'].astype('str')
df = df.merge(df2,on='UNITID', how = 'outer', indicator='True')
```

```python
df = df[df['True']!='both']
```

```python
#trying fuzzymatch with recordLinkage 
import recordlinkage

indexer = recordlinkage.Index()
indexer.full()
dupe_links = indexer.index(df,df2)

compare_cl = recordlinkage.Compare()

compare_cl.string('INSTNM', 'School', method='jarowinkler', threshold=0.93, label='School')

features = compare_cl.compute(dupe_links, df, df2)
```

```python
features.reset_index(inplace=True)  
features
```

```python
#remove non matches 
features = features[features.School==1]
features
```

```python
#merge to see match in insitutions 
df['level_0'] = df.index 
df2['level_1'] = df2.index 
features = features.merge(df, on='level_0')
features = features.merge(df2, on='level_1')
```

```python
# pd.set_option('display.max_rows',200)
# features[['School', 'INSTNM_x','UNITID_x']].to_csv('fuzzymatch.csv')
df3 = pd.read_csv('fuzzymatch.csv')
df3 = df3[['INSTNM_x','UNITID_x']]
df3 = df3.rename(columns={'UNITID_x':'UNITID'})
```

```python
df3.UNITID = df3.UNITID.astype('str')

df = df.merge(df3, on='UNITID', how='outer',indicator='Merged')
```

```python
df = df[df.Merged!='both']
```

```python
df = df[df['True']!='right_only']
```

```python
df = df[df.UNITID!='434584']
df = df[df.UNITID!='102711']
df = df[~df.INSTNM.str.contains('Community College')]
df = df[~df.INSTNM.str.contains('Technical College')]
df = df[~df.INSTNM.str.contains('Vocational')]
df = df[~df.INSTNM.str.contains('Military College')]
df = df[~df.INSTNM.str.contains('Adult')]
```

```python
print(df.UNITID.count())
df.head()
```

```python
df['SAT_AVG'] = df['SAT_AVG'].fillna(0)
df['SAT_AVG_1'] = df['SAT_AVG_1'].apply(lambda x:1 if x!=0 else 0)

top_Instituition = (df['SAT_AVG_1'].value_counts()*100/2960)
ax = top_Instituition[0:2].plot(kind='barh', color ='#91B9D6')
plt.xlabel('No. of Individuals')
plt.title('Unmatched Records - SAT')
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
```

```python
df['ACTCMMID'] = df['ACTCMMID'].fillna(0)
df['ACT_1'] = df['ACTCMMID'].apply(lambda x:1 if x!=0 else 0)

top_Instituition = (df['ACT_1'].value_counts()*100/2960)
ax = top_Instituition[0:2].plot(kind='barh', color ='#897C90')
plt.title('Remaining Unmatched Records - ACT')
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
```

Note - There remain more non ACT than Non-SAT Records. 

```python
#df.to_csv('universities_data_semi_cleaned.csv')
print('Universities without Community Colleges and For-Profit Universities are now saved under data folder of shared drive')
```

```python
df['ADM_RATE_ALL_1'] = df['ADM_RATE_ALL']
df['ADM_RATE_ALL_1'] = df['ADM_RATE_ALL_1'].fillna(0)
df['ADM_RATE_ALL_1'] = df['ADM_RATE_ALL_1'].apply(lambda x:1 if x!=0 else 0)
df['ADM_RATE_ALL_1'] = df['ADM_RATE_ALL_1'].apply(lambda x:'Present' if x!=0 else 'Missing')

top_Instituition = (df['ADM_RATE_ALL_1'].value_counts()*100/2960)
ax = top_Instituition[0:2].plot(kind='barh', color ='#202E4E')
plt.title('Remaining Unmatched Records - Admission Rate')
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
```
